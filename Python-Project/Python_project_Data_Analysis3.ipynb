```python
party = a.groupby(['Party'], as_index=False)['Total Votes'].sum().sort_values(by='Total Votes', ascending=False).head(10)
party
```

```python
plt.bar(party["Party"], party["Total Votes"])
plt.xticks(rotation="vertical")
plt.xlabel("Party")
plt.ylabel("Total Votes")
plt.title("Top 10 parties with heighest votes")
```

```python
sns.barplot(x="Party", y="Total Votes",data=party)
plt.xticks(rotation="vertical")
plt.show
```

```python
party_evmvotes = a.groupby(['Party'], as_index=False)['EVM Votes'].sum().sort_values(by='EVM Votes', ascending=False).head(15)
party_evmvotes
```

```python
plt.figure(figsize=(25,7))
result = sns.barplot(x="Party", y="EVM Votes", data=party_evmvotes, color='Green')
result.bar_label(result.containers[0])

plt.title("Top 15 parties with there EVM votes")
plt.xticks(rotation='vertical')
plt.show()
```

```python
party_result = a.groupby(['State'], as_index=False)['Total Votes'].sum().sort_values(by='Total Votes', ascending=False).head(10)
party_result
```

```python
# To find the total votes of bjp in major state
```

```python
bjp = a[a['Party'] == "Bharatiya Janata Party"]
bjp
```

```python
bjp_state = bjp.groupby(['State'], as_index=False)['Total Votes'].sum().sort_values(by='Total Votes', ascending=False).head(10)
bjp_state
```

```python
plt.figure(figsize=(20, 10))
wedges, texts, autotexts = plt.pie(
    bjp_state["Total Votes"],
    labels=bjp_state['State'],
    autopct="%1.2f%%",
    shadow=True,
    startangle=140,
    pctdistance=0.85,
    explode=[0.1] * len(bjp_state),
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}
)
plt.legend(wedges, bjp_state['State'], title="States", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.axis('equal')
plt.title("Top 10 states with higher voting % for BJP")
plt.show()
```

```python
cong = a[(a['Party'] == 'Indian National Congress')]
cong_state = cong.groupby(['State'], as_index=False)['Total Votes'].sum().sort_values(by='Total Votes', ascending=False).head(15)
cong_state
```

```python
plt.figure(figsize=(10, 5))
plt.bar(bjp_state['State'], bjp_state['Total Votes'], color='pink')
plt.bar(cong_state['State'], cong_state['Total Votes'], color='blue')
plt.legend(['BJP', 'INC'])
plt.xlabel("States")
plt.ylabel("Total Votes")
plt.title("Top 15 States with highest vote share of BJP AND INC")
plt.xticks(rotation='vertical')
plt.show()
```

```python
vote_pr = a.groupby(['Party','State'], as_index=False)['% of Votes'].sum().sort_values(by='% of Votes', ascending=False).head(50)
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
vote_pr
```

```python
pivot_df = vote_pr.pivot_table(index='State', columns='Party', values='% of Votes', fill_value=0)

# Plot the stacked bar chart
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(14, 8))

# Add labels and title
plt.xlabel('State')
plt.ylabel('% of Votes')
plt.title(' Top 50 Parties with there votes in each states')
plt.legend(title='Party', bbox_to_anchor=(1.05, 1), loc='upper left')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# To find total number of constituency in each state
```

```python

constituency_counts = a.groupby('State')['Constituency'].nunique().reset_index()
constituency_counts.rename(columns={'Constituency': 'Total Constituencies'}, inplace=True)
constituency_counts
```

```python
plt.figure(figsize=(15,10))
plt.stem(constituency_counts['State'], constituency_counts['Total Constituencies'], linefmt ='black', markerfmt ='D')
plt.title("Total number of constituencies in each state")
plt.xticks(rotation=90)
plt.show()
```
