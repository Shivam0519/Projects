```python
# Exploratory Data analysis
```

```python
group1 = data.groupby(["Gender"])
group1["Amount"].sum()
```

```python
data.head()
```

```python
sales_gender = data.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gender
```

```python
plt.bar(sales_gender["Gender"], sales_gender["Amount"])
plt.xlabel("gender")
plt.ylabel("Amount")
plt.title("amount spend by gender")
```

```python
sns.barplot("Gender", "Amount", data=sales_gender)
```

```python
sales_age = data.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_age
```

```python
plt.figure(figsize=(8,5))
sns.barplot("Age Group", "Amount", data=sales_age).set(title="Highest to lowest amount spend by age_grop")
plt.show()
```

```python
sales_state = data.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
```

```python
sales_state
```

```python
plt.figure(figsize=(25,7))
sns.barplot("State", "Orders",data=sales_state).set(title="Total number of orders statewise")
plt.show()
```

```python
sales_state_amount = data.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
```

```python

sales_state_amount
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_state_amount["Amount"],labels=sales_state_amount['State'],autopct="%0.1f%%")
plt.show()
```

```python
data.head()
```

```python
sales_occupation = data.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_occupation
```

```python
plt.figure(figsize=(10,7))
sns.barplot("Occupation", "Amount", data=sales_occupation).set(title="Top 10 Occupations spended amount on diwali")
plt.show()
```

```python
sales_catg = data.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_catg
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_catg["Amount"],labels=sales_catg['Product_Category'],autopct="%0.1f%%")
plt.show()
```

```python
Food_sales = data[data["Product_Category"] == "Food"].groupby(["Product_Category","Zone"], as_index=False).sum()
Food_sales
                                                                                                                             
```

```python
plt.figure(figsize=(8,8))
plt.pie(Food_sales["Amount"],labels=Food_sales['Zone'],autopct="%0.1f%%")
plt.show()
```
