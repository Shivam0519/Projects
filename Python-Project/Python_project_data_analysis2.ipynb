```python
group = data.groupby('year_of_release')
group.sum()
```

```python
#Top 5 imdb_votes acording to there year
data.groupby('year_of_release')['imdb_votes'].sum().sort_values(ascending=False).head(5)
```

```python
#First element from each grouped year.
group.first()
```

```python
group1 = group.describe()
group1
```

```python
years = group1.index
min_ratings = group1['imdb_rating']['min']
max_ratings = group1['imdb_rating']['max']

plt.plot(years, min_ratings, marker='o', linestyle='-', color='b', label='Min IMDb Rating')
plt.plot(years, max_ratings, marker='o', linestyle='-', color='r', label='Max IMDb Rating')

plt.xlim(2001,2019)
plt.xlabel('Year of Release')
plt.ylabel('IMDb Rating')
plt.title('Min and Max IMDb Ratings Over Years')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
